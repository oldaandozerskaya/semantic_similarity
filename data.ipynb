```python
path='dataset_rus.txt'
handle=open(path,'r')

def make_pairs(sentences):
    pairs=[]
    for i in range(len(sentences)):
        for j in range(i+1, len(sentences)):
            pairs.append([sentences[i], sentences[j]])
    return pairs

sentences=[]
i=0

for line in handle:
    #try:
        if len(line)>1:
            pos=line.find('	')
            num=int(line[pos-1:pos])
            sent=line[pos+1:len(line)-2].replace('\n','').replace(' .','')
            #json=translate_me(sent)
            #sentences.append(''.join(json["text"]))
            sentences.append(sent)

            if num==4:
                i+=1

                #делаем пары
                pairs=make_pairs(sentences)

                #в файл
                df=pd.DataFrame(pairs)
                with open('pairs_rus1.csv', 'a') as f:
                    df.to_csv(f, sep=';', index=False, header=False)

                sentences=[] 
```
